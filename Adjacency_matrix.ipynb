```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def feedforward(adjacency_matrix, inputs, weights):
    layer_input = np.dot(adjacency_matrix, inputs)  # Tính toán đầu vào của mỗi nơ-ron
    layer_output = sigmoid(np.dot(weights, layer_input))  # Tính đầu ra
    return layer_output

# Định nghĩa ma trận kề (Adjacency Matrix)
adjacency_matrix = np.array([
    [0, 1, 1],
    [0, 0, 1],
    [0, 0, 0]
])

# Trọng số giữa các nơ-ron
weights = np.array([
    [0.5, 0.8, 0.2],
    [0.3, 0.7, 0.5],
    [0.6, 0.9, 0.4]
])

# Đầu vào của mạng nơ-ron
inputs = np.array([1, 0.5, 0.2])

# Chạy quá trình lan truyền tiến
output = feedforward(adjacency_matrix, inputs, weights)
print("Output của mạng nơ-ron:", output)

# Biểu diễn đồ thị
G = nx.DiGraph()
num_nodes = adjacency_matrix.shape[0]
for i in range(num_nodes):
    for j in range(num_nodes):
        if adjacency_matrix[i, j] == 1:
            G.add_edge(f"Neuron {i+1}", f"Neuron {j+1}")

plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=10, font_weight='bold', arrows=True)
plt.title("Biểu diễn đồ thị của mạng nơ-ron")
plt.show()

```
