```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Tạo ma trận liên kết (adjacency matrix) cho mạng nơ-ron
# Mỗi hàng thể hiện các kết nối từ một nơ-ron tới các nơ-ron khác
adjacency_matrix = np.array([
    [0, 1, 1],  # Nơ-ron 1 kết nối với nơ-ron 2 và 3
    [0, 0, 1],  # Nơ-ron 2 kết nối với nơ-ron 3
    [0, 0, 0]   # Nơ-ron 3 không kết nối với nơ-ron nào
])

# Tính ma trận bậc (Degree matrix)
degree_matrix = np.diag(np.sum(adjacency_matrix, axis=1))

print("Ma trận liên kết (Adjacency Matrix):")
print(adjacency_matrix)
print("\nMa trận bậc (Degree Matrix):")
print(degree_matrix)

# Tạo đồ thị từ ma trận liên kết
G = nx.DiGraph()

# Thêm các cạnh từ ma trận liên kết
for i in range(adjacency_matrix.shape[0]):
    for j in range(adjacency_matrix.shape[1]):
        if adjacency_matrix[i, j] == 1:
            G.add_edge(f"Neuron {i+1}", f"Neuron {j+1}")

# Vẽ đồ thị
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G)  # Cách sắp xếp đồ thị
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', arrows=True)
plt.title("Đồ thị mạng nơ-ron")
plt.show()
```
